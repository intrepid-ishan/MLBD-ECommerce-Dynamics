#### 3.1.1 Regional EDA

<p>
    <strong>Regional Analysis of E-commerce Data</strong><br>
    In this section, we delve into a regional analysis of e-commerce activities in Brazil, leveraging the first three digits of zip codes to segment data geographically. Our analysis aims to uncover regional trends and disparities in several key metrics:
</p>
<ul>
    <li><strong>Total Revenue:</strong> By analyzing total revenue by region, we can identify areas with higher economic activity and customer spending. This can reveal whether revenue is concentrated in major urban areas or more evenly distributed across the country.</li>
    <li><strong>Average Order Price:</strong> Investigating the average order price helps us understand regional purchasing power and consumer behavior. For instance, higher average prices in remote areas might indicate bulk purchasing due to less frequent shopping trips or limited access to stores.</li>
    <li><strong>Freight Ratio:</strong> The freight ratio gives us insights into the logistical costs associated with different regions. A higher freight ratio in remote areas could suggest increased transportation costs due to the distance from major distribution centers.</li>
    <li><strong>Average Delivery Time:</strong> Examining the average delivery time across regions helps us assess the efficiency and challenges of logistics networks. Longer delivery times in certain areas might be indicative of logistical complexities or infrastructural limitations.</li>
</ul>
<p>
    The decision to use the <b>first three digits of zip codes as a regional marker</b> allows for a balanced approach between granularity and manageability. This level of segmentation is specific enough to differentiate between key regions (urban, suburban, rural) while avoiding the complexities and data sparsity that might arise with more granular divisions such as full zip codes.
</p>

```python
import geopandas as gpd
import matplotlib.pyplot as plt

def plot_metric_by_zip(gdf, metric_column, title, color='blue', figsize=(12, 12)):
    # Load the map of Brazil
    world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
    brazil = world[world.name == "Brazil"]

    # Plotting
    fig, ax = plt.subplots(1, 1, figsize=figsize)
    brazil.plot(ax=ax, color='white', edgecolor='black')

    # Calculate marker sizes, avoid division by zero
    max_value = gdf[metric_column].max()
    if max_value == 0:
        max_value = 1  # Prevent division by zero
    marker_sizes = (gdf[metric_column] / max_value) * 100

    # Plot each 3-digit prefix's metric as a point on the map
    gdf.plot(ax=ax, marker='o', color=color, markersize=marker_sizes, legend=True)

    plt.title(title)
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.show()
```

```python
# Convert the geolocation_zip_code_prefix to string and extract the first 3 digits
geolocation['zip_code_prefix_3'] = geolocation['geolocation_zip_code_prefix'].astype(str).str[:3]

# Group by the 3-digit prefix to find the mean latitude and longitude for each prefix
geo_avg_coords = geolocation.groupby('zip_code_prefix_3').agg({
    'geolocation_lat': 'mean',
    'geolocation_lng': 'mean'
}).reset_index()
```

<p />

<strong>Total Revenue by 3-Digit Zip Code Prefix in Brazil</strong>
<p>
    The visualization of total revenue by the 3-digit zip code prefix across Brazil highlights a distinct concentration of economic activity. A significant portion of revenue is generated in the southern and southeastern regions of the country, aligning with the more densely populated and economically developed areas. The map showcases a cluster of high-revenue areas along the coastline, reflecting the bustling urban centers such as São Paulo and Rio de Janeiro. The concentration of green markers in these regions illustrates the vitality of commerce, likely attributable to a combination of population density, higher income levels, and a well-established infrastructure.
</p>

```python
# Calculate the total revenue for each 3-digit prefix
total_revenue_by_zip_prefix = merged_orders.groupby('zip_code_prefix_3')['price'].sum().reset_index()
total_revenue_by_zip_prefix.rename(columns={'price': 'total_revenue'}, inplace=True)

# Merge total revenue data with the coordinates
geo_total_revenue = pd.merge(total_revenue_by_zip_prefix, geo_avg_coords, on='zip_code_prefix_3', how='inner')

# Prepare the GeoDataFrame for total revenue
gdf_total_revenue = gpd.GeoDataFrame(
    geo_total_revenue,
    geometry=gpd.points_from_xy(geo_total_revenue.geolocation_lng, geo_total_revenue.geolocation_lat)
)

# Plot the total revenue by 3-digit zip code prefix
plot_metric_by_zip(gdf_total_revenue, 'total_revenue', 'Total Revenue by 3-Digit Zip Code Prefix in Brazil', color='green')
```

<p />

<strong>Average Order Price by 3-Digit Zip Code Prefix in Brazil</strong>
<p>
    This map illustrates the average order price distribution across different regions in Brazil, denoted by the 3-digit zip code prefix. Notably, there is a prominent circle in the north-western area, suggesting a higher average order price in that particular zip code. While the southern and southeastern regions show a multitude of smaller circles, indicating a lower average order price in these populous and commercially active areas. This could suggest more competitive pricing or more frequent, but smaller purchases in the more affluent regions. Alternatively, the larger circles scattered across the map might reflect less frequent purchases or bulk buying behaviors, potentially influenced by varying freight costs, which will be examined subsequently.
</p>

```python
# Calculate the average order price for each 3-digit prefix
average_price_by_zip_prefix = merged_orders.groupby('zip_code_prefix_3')['price'].mean().reset_index()
average_price_by_zip_prefix.rename(columns={'price': 'avg_price'}, inplace=True)

# Merge average price data with the coordinates
geo_avg_price = pd.merge(average_price_by_zip_prefix, geo_avg_coords, on='zip_code_prefix_3', how='inner')

# Step 4: Plot the results
import matplotlib.pyplot as plt
import geopandas as gpd

# Convert to GeoDataFrame for plotting
gdf_avg_price = gpd.GeoDataFrame(
    geo_avg_price,
    geometry=gpd.points_from_xy(geo_avg_price.geolocation_lng, geo_avg_price.geolocation_lat)
)

# Plot the average order price by 3-digit zip code prefix
plot_metric_by_zip(gdf_avg_price, 'avg_price', 'Average Order Price by 3-Digit Zip Code Prefix in Brazil', color='blue')
```

<p />

<strong>Average Freight Ratio by 3-Digit Zip Code Prefix in Brazil</strong>
<p>
    The distribution of average freight ratios across Brazilian zip codes reveals a notable pattern. In the southeastern region, a dense clustering of points suggests a higher frequency of deliveries. This could be a reflection of the urban density and higher volume of e-commerce transactions in these areas. On the other hand, the noticeably larger circles in more remote northern regions could indicate areas where deliveries are less frequent but more costly on a per-item basis. ```python
# Calculate the freight ratio for each order
merged_orders['freight_ratio'] = merged_orders['freight_value'] / merged_orders['price']

# Calculate the average freight ratio for each 3-digit prefix
average_freight_ratio_by_zip_prefix = merged_orders.groupby('zip_code_prefix_3')['freight_ratio'].mean().reset_index()
average_freight_ratio_by_zip_prefix.rename(columns={'freight_ratio': 'avg_freight_ratio'}, inplace=True)

# Merge average freight ratio data with the coordinates
geo_avg_freight_ratio = pd.merge(average_freight_ratio_by_zip_prefix, geo_avg_coords, on='zip_code_prefix_3', how='inner')

# Convert to GeoDataFrame for plotting
gdf_avg_freight_ratio = gpd.GeoDataFrame(
    geo_avg_freight_ratio,
    geometry=gpd.points_from_xy(geo_avg_freight_ratio.geolocation_lng, geo_avg_freight_ratio.geolocation_lat)
)

# Plot the average freight ratio by 3-digit zip code prefix
plot_metric_by_zip(gdf_avg_freight_ratio, 'avg_freight_ratio', 'Average Freight Ratio by 3-Digit Zip Code Prefix in Brazil', color='purple')
```

<strong>Average Delivery Time by 3-Digit Zip Code Prefix in Brazil</strong>
<p>
    The map of average delivery times in Brazil illustrates a distinct regional disparity. In the bustling southeast, a profusion of smaller markers signifies shorter delivery periods, which likely benefits from well-established logistics networks in these densely populated urban centers. ```python
# Calculate the actual delivery time
merged_orders['order_delivered_customer_date'] = pd.to_datetime(merged_orders['order_delivered_customer_date'])
merged_orders['order_delivered_carrier_date'] = pd.to_datetime(merged_orders['order_delivered_carrier_date'])
merged_orders['actual_delivery_time'] = (merged_orders['order_delivered_customer_date'] -
                                         merged_orders['order_delivered_carrier_date']).dt.days

# Create a copy of the merged_orders DataFrame for delivery time analysis
delivery_time_orders = merged_orders.copy()

# Clean the data by removing negative delivery times and replacing NaN values
delivery_time_orders = delivery_time_orders[delivery_time_orders['actual_delivery_time'] >= 0]

# Calculate the average delivery time for each 3-digit prefix
average_delivery_time_by_zip_prefix = delivery_time_orders.groupby('zip_code_prefix_3')['actual_delivery_time'].mean().reset_index()
average_delivery_time_by_zip_prefix.rename(columns={'actual_delivery_time': 'avg_delivery_time'}, inplace=True)

# Merge average delivery time data with the coordinates
geo_avg_delivery_time = pd.merge(average_delivery_time_by_zip_prefix, geo_avg_coords, on='zip_code_prefix_3', how='inner')

# Convert to GeoDataFrame for plotting
gdf_avg_delivery_time = gpd.GeoDataFrame(
    geo_avg_delivery_time,
    geometry=gpd.points_from_xy(geo_avg_delivery_time.geolocation_lng, geo_avg_delivery_time.geolocation_lat)
)

# Plot the average delivery time by 3-digit zip code prefix
plot_metric_by_zip(gdf_avg_delivery_time, 'avg_delivery_time', 'Average Delivery Time by 3-Digit Zip Code Prefix in Brazil', color='orange')
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

# Custom transformer for adding new attributes
class FeaturesAdder(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        transformed_df = X.copy()

         # Convert date columns to datetime
        date_columns = ['order_purchase_timestamp', 'order_approved_at', 'order_estimated_delivery_date', 'order_delivered_customer_date']
        for col in date_columns:
            transformed_df[col] = pd.to_datetime(transformed_df[col], errors='coerce')

        # Apply the date calculations
        transformed_df['working_days_est_delivery'] = transformed_df.apply(
            lambda x: cal.get_working_days_delta(x['order_approved_at'], x['order_estimated_delivery_date'])
            if pd.notnull(x['order_approved_at']) and pd.notnull(x['order_estimated_delivery_date'])
            else None, axis=1)

        transformed_df['working_days_actual_delivery'] = transformed_df.apply(
            lambda x: cal.get_working_days_delta(x['order_approved_at'], x['order_delivered_customer_date'])
            if pd.notnull(x['order_approved_at']) and pd.notnull(x['order_delivered_customer_date'])
            else None, axis=1)

        transformed_df['working_days_delivery_delta'] = transformed_df['working_days_actual_delivery'] - transformed_df['working_days_est_delivery']
        transformed_df['is_delivery_late'] = transformed_df['order_delivered_customer_date'] > transformed_df['order_estimated_delivery_date']
        transformed_df['avg_product_value'] = transformed_df['price'] / transformed_df['order_items_qty']
        transformed_df['total_order_value'] = transformed_df['price'] + transformed_df['freight_value']
        transformed_df['freight_ratio'] = transformed_df['freight_value'] / transformed_df['price']
        transformed_df['purchase_weekday'] = transformed_df['order_purchase_timestamp'].dt.dayofweek

        # Select columns to keep
        columns_to_keep = [
            'working_days_est_delivery',
            'working_days_actual_delivery',
            'working_days_delivery_delta',
            'is_delivery_late',
            'avg_product_value',
            'total_order_value',
            'freight_ratio',
            'purchase_weekday',
            'customer_state',
            'seller_state',
            'product_category_name',
            'class_label'  # Keep the label for splitting
        ]
        transformed_df = transformed_df[columns_to_keep]

        return transformed_df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# Separate features and labels
X = feat_eng_df.drop('class_label', axis=1)
y = feat_eng_df['class_label']

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Numeric features to be scaled
numeric_features = ['working_days_est_delivery', 'working_days_actual_delivery', 'working_days_delivery_delta',
                    'avg_product_value', 'total_order_value', 'freight_ratio', 'purchase_weekday']
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

# Categorical features to be one-hot encoded
categorical_features = ['customer_state', 'seller_state', 'product_category_name']
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combining transformers into a preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
])

# Applying preprocessing
X_train_preprocessed = preprocessor.fit_transform(X_train)
X_test_preprocessed = preprocessor.transform(X_test)
```

### 3.2 Model training and evaluation

<p>For the task of predicting customer review sentiment, a variety of machine learning models were trained and evaluated. Each model was chosen based on its ability to handle binary classification tasks and its different approach to learning from the data.</p>

<ul>
  <li><strong>Logistic Regression:</strong> A foundational statistical method that estimates probabilities using a logistic function. Chosen for its simplicity and interpretability.</li>
  <li><strong>Decision Tree:</strong> A flowchart-like tree structure where an internal node represents a feature, the branch represents a decision rule, and each leaf node represents the outcome. Chosen for its ease of understanding and ability to capture non-linear relationships.</li>
  <li><strong>Random Forest:</strong> An ensemble of decision trees, typically trained with the "bagging" method. Chosen for its robustness and better generalization.</li>
  <li><strong>XGBoost:</strong> An implementation of gradient boosted decision trees designed for speed and performance. Chosen for its effectiveness in dealing with various types of structured data.</li>
</ul>

<p>The primary evaluation metric selected is the <b>F1 score</b>, which harmonizes the precision and recall of the classifier into a single metric. It is particularly useful when the class distribution is imbalanced, ensuring that both the proportion of actual positives correctly identified and the relevance of the model's predictions are considered.

#### 3.2.1 Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Initialize Logistic Regression model
log_reg = LogisticRegression(max_iter=1000, random_state=42)

# Train the model
log_reg.fit(X_train_preprocessed, y_train)

# Predictions
y_pred = log_reg.predict(X_test_preprocessed)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print("Accuracy:", accuracy)
print("Classification Report:\n", class_report)
```

<p>The Logistic Regression model serves as our initial benchmark. While it achieved a relatively high overall accuracy of approximately 78.84%, its performance on Class 0 (negative reviews) was suboptimal. Specifically, the model showed a low recall for Class 0, suggesting it was not effective at correctly identifying negative reviews. This limited performance highlights the need for more sophisticated models to better handle the nuances of our dataset. The evaluation of Logistic Regression sets the stage for comparison with more complex models to follow.</p>

#### 3.2.2 Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

# Initialize the Decision Tree model
dt_classifier = DecisionTreeClassifier(random_state=42)

# Train the model on the training data
dt_classifier.fit(X_train_preprocessed, y_train)

# Make predictions on the testing data
y_pred_dt = dt_classifier.predict(X_test_preprocessed)

# Calculate accuracy and classification report
accuracy_dt = accuracy_score(y_test, y_pred_dt)
class_report_dt = classification_report(y_test, y_pred_dt)

print("Decision Tree Accuracy:", accuracy_dt)
print("Decision Tree Classification Report:\n", class_report_dt)
```

<p>The Decision Tree model marked a notable improvement from the Logistic Regression model, especially in terms of its ability to identify negative reviews (Class 0). It achieved an overall accuracy of approximately 75.81% and demonstrated a more balanced performance across both classes. The model's recall for Class 0 increased significantly, indicating better identification of negative reviews compared to the Logistic Regression model. This improvement in performance suggests the potential benefits of more complex models like Random Forest, which builds upon the decision tree approach by integrating multiple trees for more robust predictions.</p>


#### 3.2.3 Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize Random Forest model
rf_classifier = RandomForestClassifier(random_state=42)

# Train the model
rf_classifier.fit(X_train_preprocessed, y_train)

# Predictions
y_pred_rf = rf_classifier.predict(X_test_preprocessed)

# Evaluate the model
accuracy_rf = accuracy_score(y_test, y_pred_rf)
class_report_rf = classification_report(y_test, y_pred_rf)

print("Random Forest Accuracy:", accuracy_rf)
print("Random Forest Classification Report:\n", class_report_rf)
```

<p>The Random Forest model marks a significant improvement over the previous models, achieving an impressive accuracy of approximately 84.23%. This model demonstrates a robust ability in accurately identifying positive reviews (Class 1), as evidenced by the high precision and recall values. Additionally, the F1 score, which is crucial for our analysis due to class imbalance, is notably better with a macro average of around 0.73. This improvement in F1 score compared to both Logistic Regression and Decision Tree models underscores the Random Forest model's enhanced capability in balancing precision and recall, especially for positive review predictions. The model's overall performance indicates a strong predictive power, making it a valuable tool in understanding customer satisfaction dynamics. As this model has exhibited the best performance thus far, we will proceed to examine its feature importances to uncover the key factors driving these predictions.</p>

```python
scale_pos_weight = len(y_train[y_train == 0]) / len(y_train[y_train == 1])


from sklearn.model_selection import GridSearchCV

# Parameters for grid search
param_grid = {
    'max_depth': [3, 6, 10],
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.3]
}

# Initialize the XGBClassifier with scale_pos_weight
xgb_clf = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', scale_pos_weight=scale_pos_weight, random_state=42)

# Grid search with cross-validation
grid_search = GridSearchCV(xgb_clf, param_grid, cv=3, scoring='f1', verbose=2)
grid_search.fit(X_train_preprocessed, y_train)

# Best parameters
print("Best parameters:", grid_search.best_params_)

# Train the model with the best parameters
best_xgb = grid_search.best_estimator_
best_xgb.fit(X_train_preprocessed, y_train)

# Predictions and evaluation
y_pred_best_xgb = best_xgb.predict(X_test_preprocessed)
accuracy_best_xgb = accuracy_score(y_test, y_pred_best_xgb)
class_report_best_xgb = classification_report(y_test, y_pred_best_xgb)

print("Best XGBoost Accuracy:", accuracy_best_xgb)
print("Best XGBoost Classification Report:\n", class_report_best_xgb)
```

<p>After conducting a grid search to optimize the XGBoost hyperparameters, we identified a set of parameters that improved the f1-score for the negative class (0), but the overall accuracy decreased to approximately 78%. <p>Although the grid search enhanced the balance between precision and recall for individual classes, it did not lead to a superior overall model compared to the Random Forest. <b>Addressing Class Imbalance with SMOTE:</b>
<p>Given the imbalanced nature of our classes, with a higher prevalence of positive reviews over negative ones, we must consider the potential bias this introduces to our model training. To mitigate this issue and strive for a model that performs well across both classes, we will employ the Synthetic Minority Over-sampling Technique (SMOTE).</p>
<p>SMOTE generates synthetic samples for the minority class, aiming to provide a more balanced distribution and thus, a fairer playground for our models to learn from. By enriching our dataset in this manner, we anticipate an improvement in the model's ability to generalize and thus, an enhancement in its predictive performance, especially for the underrepresented class.</p>
<p>We will re-train XGBoost and Random Forest models after SMOTE, since these two have performed the best so far.</p>


```python
from imblearn.over_sampling import SMOTE

# Define SMOTE
smote = SMOTE(random_state=42)

# Apply SMOTE to the training data after preprocessing
X_train_preprocessed = preprocessor.fit_transform(X_train)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_preprocessed, y_train)
```
